```python
#fran_list 각 요소를 포함하는 main_final의 인덱스 리턴 
def fr(i):
  try:
    a=main_final['사업장명'].str.contains(i)
    if a.sum()>0:
      return np.where(a==True)
  except:
    pass
```

```python
#해당 리턴값들 '여부'라는 새로운 칼럼에 저장
fran_list['여부'] = fran_list['상호'].map(lambda x : fr(x))
```

```python
#'여부'칼럼의 값들 빈 리스트에 모두 추가
f_list=[]
for i in range(len(fran_list)):
  try:
    for j in range(len(fran_list['여부'][i][0])):
      f_list.append(fran_list['여부'][i][0][j])
  except:
    pass  
```

```python
#set로 중복제거하고 가맹여부에 해당하는 인덱스만 추출
f_list
```

```python
#f_list에 들어있는 친구들 가맹여부 1로 변환
for i in range(len(f_list)):
  main_final['가맹여부'][f_list[i]]=1
```

## **외식업종류 정리, 연/분기 칼럼생성**

```python
#분류해야할 원데이터의 업태명들
set(main_final['위생업태명'])
```

```python
#분류기준
sector = ['한식음식점','중식음식점','일식음식점','양식음식점','제과점','패스트푸드점','치킨전문점','분식전문점','호프간이주점','커피음료','기타']
```

```python
#'업종분류'라는 새로운 칼럼 생성
main_final['업종분류'] = '기타'
```

```python
#if문을 이용한 업종분류
for i in range(len(main_final)):
  if main_final['위생업태명'][i] in ['냉면집','탕류(보신용','한식','횟집']:
    main_final['업종분류'][i]='한식음식점'
  elif main_final['위생업태명'][i] == '중국식':
    main_final['업종분류'][i]='중식음식점'
  elif main_final['위생업태명'][i] in ['복어취급','일식']:
    main_final['업종분류'][i]='일식음식점'
  elif main_final['위생업태명'][i] in ['경양식','패밀리레스토랑']:
    main_final['업종분류'][i]='양식음식점'
  elif main_final['위생업태명'][i] == '패스트푸드':
    main_final['업종분류'][i]='패스트푸드점'
  elif main_final['위생업태명'][i] == '통닭(치킨)':
    main_final['업종분류'][i]='치킨전문점'
  elif main_final['위생업태명'][i] in ['김밥(도시락)','분식']:
    main_final['업종분류'][i]='분식전문점'
  elif main_final['위생업태명'][i] in ['호프/통닭','정종/대포집/소주방','감성주점']:
    main_final['업종분류'][i]='호프간이주점'
  elif main_final['위생업태명'][i] == '까페':
    main_final['업종분류'][i]='커피음료'
  else:
    main_final['업종분류'][i]='기타'
```
