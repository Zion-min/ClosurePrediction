상권별 매출 금액, 건수는 상권내 점포 전체 합산에 해당
=> 해당 변수는 점포수로 나눠주어야 함


```python
nulind=kwan.loc[:,'점포수_x']!=0
```

```python
nulind
```

```python
kwan.loc[nulind,'점포당_매출_금액_x']=kwan.loc[nulind,'분기당_매출_금액_x']/kwan.loc[nulind,'점포수_x']
kwan.loc[nulind,'점포당_매출_건수_x']=kwan.loc[nulind,'분기당_매출_건수_x']/kwan.loc[nulind,'점포수_x']
kwan.loc[nulind,'건수당_매출_금액_x']=kwan.loc[nulind,'분기당_매출_금액_x']/kwan.loc[nulind,'분기당_매출_건수_x']
```

## **이상치**

```python
def get_bound(data):
  q1 = np.percentile(data,25)
  q3 = np.percentile(data,75)

  iqr = q3-q1
  lower_bound = q1-(iqr*1.5)
  upper_bound = q3+(iqr%1.5)

  return lower_bound, upper_bound

def get_outlier(data):
  lower_bound, upper_bound = get_bound(data)
  return data[np.where((data>upper_bound)|(data<lower_bound))]
```
